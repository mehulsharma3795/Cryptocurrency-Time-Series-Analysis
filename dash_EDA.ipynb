```python
# The test statistic is way more than the critical statistics threfore the time series is not stationary. Obviously we can see that
# Neither the mean is constant nor statndard devaition.
```

```python
# Estiminating an eliminating trend and seasonality using Diffrrencing
# Returns in our dataframe is actually diffrence only.
# By running the stationarity test on Bitcoin returns we found that this time series is stationary
rolmean = df_dash.data.Returns.rolling(30).mean()
rolstd = df_dash.data.Returns.rolling(30).std()
plt.plot(df_dash.data.Returns,color='blue',label='Original')
plt.plot(rolmean,color='red',label='Rolling Mean')
plt.plot(rolstd,color='black',label='Rolling Std')
plt.legend(loc='best')
plt.title("Rolling Mean & Standard Deviation")
plt.show()
# Perform Dickey Fuller Test 
print("Results of Dickey-Fuller Test")
df_dash.data.Returns.dropna(inplace=True)
dftest = adfuller(df_dash.data.Returns)
dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,Value in dftest[4].items():
    dfoutput['Critical Value(%s)'%key] = Value
print(dfoutput)
print ('Results of KPSS Test:')
print("------------------------------------------------------------------------------")
kpsstest = kpss(df_dash.data.Returns, regression='c')
kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
for key,value in kpsstest[3].items():
    kpss_output['Critical Value (%s)'%key] = value
print (kpss_output)
print("------------------------------------------------------------------------------")
print ('Results of Phillips-Perron Test:')
pptest = PhillipsPerron(df_dash.data.Returns)
print(pptest)
print("------------------------------------------------------------------------------")
```

```python
from statsmodels.tsa.stattools import acf,pacf
lag_acf = acf(df_dash.data.Returns,nlags=40,fft=False)
lag_pacf = pacf(df_dash.data.Returns,nlags=40)
```

```python
# Plotting ACF and PACF plots
plt.figure(figsize=(20,8))
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_dash.data.Returns)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_dash.data.Returns)),linestyle='--',color='gray')
plt.title("Autocorrelation Plot")
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_dash.data.Returns)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_dash.data.Returns)),linestyle='--',color='gray')
plt.title("Partial Autocorrelation Plot")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(14,7))
acf = smt.graphics.plot_acf(df_dash.data.Returns, lags=40 , alpha=0.05,ax=ax)
plt.title("Autocorrelation Plot")
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(14,7))
acf = smt.graphics.plot_pacf(df_dash.data.Returns, lags=40 , alpha=0.05,ax=ax)
plt.title("Partial Autocorrelation Plot")
plt.show()
```

```python
# Value of p - where the PACF chart crossses the upper confidence level for the first time p=1
# Value of q - where the ACF chart crossses the upper confidence level for the first time  q=1
# Since the time Series has both AR & MA terms therefore we'll not build an AR or MA model but an ARIMA model
```
